```python
def linear__regression(X,y,alpha=1):
  x_mean = X.mean
  y_mean =y.mean

  num = 0
  den = 0

  for i in range(X.shape[0]):
    num = num +(y[i]-y_mean) * (X[i]-x_mean)
    den = den + (X[i]-x_mean) * (X[i]-x_mean)

  m=num/(den +alpha)
  b = y_mean - m*x_mean

  return m,b
```
